# Expectancy in trading
从交易的角度来看，如果两个策略导致利润的数学期望相同，更加平稳的交易策略值得采纳。<br />
e.g. <br />
Strategy 1: 1 dollar, 100%; <br />
Strategy 2: 10 dollars, 10%; <br />

If we play 12 times, it depends with S2(10, 20,...It is more probable that you only earn 10 dollars);the profit is definitely 12 if S1.
<br />

关键在于概率是怎么算的。

## Batting average
Measuring the percentage of profitable trades out of the total number of trades.(交易10天中8天赚钱、batting average = 80%) <br />

## Win-loss ratio


```python
win_ratio = (5 + 8 + 4 + 5 + 4 + 1) / 6
loss_ratio = abs(-4-6-7-2) / 4
win_loss_ration = win_ratio / loss_ratio
# 平均值的计算会因为其中某一笔或几笔交易的原因失衡，最终要回到具体问题分析。
display(win_loss_ration)
```

Only about 0.95 is earned per loss

An example(今后自己写策略回测的时候可以遵照这个指标) <br />
In fact, 65% batting average is a good number.  An example: If Strategy 2 is changed to: 11 dollars 10%, it is better to adopt S2 with lower batting average.
